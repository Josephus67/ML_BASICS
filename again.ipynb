# Reading our data set

```python
import pandas as pd

g_train = pd.read_csv('train.csv')
g_train
```

# CHECKING FOR NULL VALUES

```python
#isnull values are age and cabin and embarked
g_train.isnull().sum()
```

# CHECKING PERCENTAGE OF NULL ENTIRES

```python
print((g_train.isnull().sum()/len(g_train))*100)
```

# CABIN HAS MORE THAN 50% OF THE ENTRIES MISSING, I'LL DROP IT

```python
#g_train = g_train.drop(columns=['Cabin'])
#g_train
g_train = g_train.drop(columns=['Cabin'])
g_train
```

# IMPUTATION ON MISSING VALUES

```python
#Age and Embarked
g_train['Embarked'].fillna(g_train['Embarked'].mode()[0], inplace=True)
g_train['Age'].fillna(g_train['Age'].median(), inplace=True)
g_train
```

```python
print((g_train.isnull().sum()/len(g_train))*100)
```

# Encoding Categorical data

```python
#Using Label Encoding for Sex and One-hot Encoding for Embarked
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
g_train['Sex'] = label_encoder.fit_transform(g_train['Sex'])

g_train = pd.get_dummies(g_train, columns=['Embarked'], prefix='Embarked')

g_train.head()
```

# Normalization of Numerical Features

```python
from sklearn.preprocessing import MinMaxScaler
#Numrical columns that can be normalized are AGE, Fare, SibSp and Parch

# Identify numerical columns
numerical_columns = ['Age', 'Fare', 'SibSp', 'Parch']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling to numerical columns
g_train[numerical_columns] = scaler.fit_transform(g_train[numerical_columns])

g_train.head()
```

# Data Visualization

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

# Survivors by Gender
survivors_by_gender = g_train.groupby('Sex')['Survived'].mean()
plt.figure(figsize=(6, 4))
survivors_by_gender.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()

# Survivors by Passenger Class
survivors_by_class = g_train.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(6, 4))
survivors_by_class.plot(kind='bar', color='purple')
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.xticks(rotation=0)
plt.show()

# Age Distribution of Survivors
plt.figure(figsize=(8, 5))
g_train[g_train['Survived'] == 1]['Age'].plot(kind='hist', bins=20, alpha=0.7, color='green', label='Survived')
g_train[g_train['Survived'] == 0]['Age'].plot(kind='hist', bins=20, alpha=0.7, color='red', label='Did Not Survive')
plt.title('Age Distribution of Survivors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Fare Distribution of Survivors
plt.figure(figsize=(8, 5))
g_train.boxplot(column='Fare', by='Survived', grid=False, patch_artist=True,
                boxprops=dict(facecolor='lightblue', color='black'),
                medianprops=dict(color='red'))
plt.title('Fare Distribution by Survival Status')
plt.suptitle('')  
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```
